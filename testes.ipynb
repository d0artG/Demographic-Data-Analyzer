```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
df = pd.read_csv("fcc-forum-pageviews.csv", index_col=0, parse_dates = True, names=["date","views"], header=0)
df
```

```python
df.drop(df[(df['views'] > df['views'].quantile(0.975)) |(df['views'] < df['views'].quantile(0.025))].index, inplace=True)
df
```

```python
fig,ax=plt.subplots()
ax=plt.plot(df.index,df["views"])
plt.title("Daily freeCodeCamp Forum Page Views 5/2016-12/2019")
plt.xlabel("Date")
plt.ylabel("Page Views")
```
